```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        if head == None:
            return head
        dub = 0
        node = head
        while node.next:
            if node.next.val == node.val:
                node.val = node.next.val
                node.next = node.next.next
                dub = 1
            else:
                if dub == 1:
                    node.val = node.next.val
                    node.next = node.next.next
                    dub = 0
                else:
                    node = node.next
                    dub = 0
        
        if dub == 1:
            node = head
            if node.next is None:
                head = None
            else:
                while node.next.next:
                    node = node.next
                node.next = None
        
        return head
```
