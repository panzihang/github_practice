```python
from typing import List

class Solution:
    def findKthPositive(self, arr: List[int], k: int) -> int:
        if arr == []:
            return k
        i=0
        num=0
        ind=0
        while i<k:
            num = num+1
            if num != arr[ind]:
                i = i+1
            else:
                ind = min(ind+1, len(arr)-1)
            # print('num:',num)
            # print('i:', i)
            # print('ind:',ind)
        
        return num
```
