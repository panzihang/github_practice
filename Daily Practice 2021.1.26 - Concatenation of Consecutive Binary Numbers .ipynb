```python
class Solution:
    def concatenatedBinary(self, n: int) -> int:
        comb_bin = ""
        for i in range(n):
            comb_bin += str(bin(i+1))[2:]
        result = int(comb_bin, 2) % (1000000000 + 7)
        return result
```

```python
sol = Solution()
```

```python
sol.concatenatedBinary(1)
```

```python
sol.concatenatedBinary(12)
```
