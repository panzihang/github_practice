```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        int1 = l1.val
        int2 = l2.val
        plus_one = 0
        lresult = ListNode()
        pointer = lresult
        while 1:
            
            # Compute
            int_sum = int1 + int2
            
            int_sum = int_sum + plus_one
            if int_sum >= 10:
                int_sum -= 10
                plus_one = 1
            else:
                plus_one = 0
            
            # Set
            pointer.val = int_sum
            
            # Update
            if (l1.next is None) and (l2.next is None):
                if plus_one == 1:
                    pointer.next = ListNode(val=1, next=None)
                    break
                else:
                    pointer.next = None
                    break
            
            if l1.next is not None:
                l1 = l1.next
                int1 = l1.val
            else:
                int1 = 0
            if l2.next is not None:
                l2 = l2.next
                int2 = l2.val
            else:
                int2 = 0

            pointer.next = ListNode(val=0, next = None)
            pointer = pointer.next
            
            
        return lresult
```

```python
l1_list = [2,5,1]
l2_list = [5,6,4]
l1 = ListNode()
l2 = ListNode()
pointer1 = l1
pointer2 = l2
for i in range(len(l1_list)):
    pointer1.val = l1_list[i]
    if i != len(l1_list)-1:
        pointer1.next = ListNode()
        pointer1 = pointer1.next
    else:
        pointer1.next = None
for j in range(len(l2_list)):
    pointer2.val = l2_list[j]
    if j != len(l2_list)-1:
        pointer2.next = ListNode()
        pointer2 = pointer2.next
    else:
        pointer2.next = None
```

```python
sol = Solution()
```

```python
lresult= sol.addTwoNumbers(l1, l2)
```
