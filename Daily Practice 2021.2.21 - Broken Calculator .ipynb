```python
class Solution:
    def brokenCalc(self, X: int, Y: int) -> int:
        i = 0
        while X != Y:
            if Y < X:
                return i + int(X - Y)
            i += 1
            if Y%2==0:
                Y = Y/2
            else:
                Y = Y + 1
                
        return i
    
```

```python
sol = Solution()
```

```python
sol.brokenCalc(2,3)
```

```python
sol.brokenCalc(3,9)
```

```python
sol.brokenCalc(6,10)
```

```python
sol.brokenCalc(1,100000000)
```
